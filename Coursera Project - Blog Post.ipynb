# Comparing Cities

The purpose of this project is to allow you to easily compare different cities based on their top venues. This is particularly useful if you are planning a holiday or moving to a new city and you want to be sure the best places in a city align with your interests. <br>

A lot of the information available about city venues is anecdotal (bloggers, travel guides etc.) and be time consuming to search through. Taking a more data-based approach you could ask what category is most popular among a cities top 100 venues? If you love parks/recreation you may be interested in London where 25 of it’s top 100 venues are parks. If you’re a foodie you may enjoy Berlin where 35 of their top 100 venues are food spots. <br>

Further to just stating numbers, it should be easy to visualize and compare how popular each category of venue is. An easy way to do this is with radar charts, if you’ve played video games such as FIFA you may recognize these when you compare player attributes. The final result should be an easy to way to compare the attributes of each city.. as if they were players on FIFA.
Utilizing Foursquare API data you can retrieve details of all top venues in a given city. Information such as Location, Rating and Category. Using category we can compare cities not only against each other but to our interests.<br>

An initial ‘out-of-the-box’ attempt looking at London’s top 100 venues shows there are far too many categories for this to be useful.


Some aggregation is required to make this more readable.

We can create distinct 8 groups to assign each category. This will then make the chart easier to read and more useful. The groups I’ve chosen are: <br>

•	Nightlife <br>
•	Shopping <br>
•	Food/Drink <br>
•	Points of intereset <br>
•	Parks <br>
•	Culture <br>
•	Entertainment <br>
•	Hotel/Accomodation <br>

By using some keywords for each category we can organize our categories. For example, keywords for Food may include: <br>

•	Restaurant <br>
•	Wings <br>
•	Café etc. <br>

If a category name contains a keyword, it is assigned to that group. Categories not contained in any group are assigned ‘Other’. 
Running this over a large list of categories across several cities we have a relatively comprehensive list of mappings from category to group <br>

Much easier to read! We can now see that of London’s top 100 venues ~20 of them are Parks. So if you enjoy spending time in Parks/Gardens then London may be the place for you! <br>
We can also compare cities using the same method, giving us an easy to compare cities or just find out which ones may appeal to our own interests.
